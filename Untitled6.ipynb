#新增檔案

```python
a=open("a.txt","w") #新增檔案
a.write("This is a test1")
a.write("This is a test2\n")
a.write("This is a test3\n")
a.close() #關閉檔案
```

#檢查檔案是否存在

```python
os.path.isfile("a.txt")
```

#讀取檔案

```python
a=open("a.txt","r")
print(a.readline())
print(a.readline())
print(a.readline()) #多餘的命令，檔案只有2行
a.close()
```

```python
a=open("a.txt","r")
print(a.readlines())  #一次讀取多行，與a.read()命令相同
a.close()
```

```python
#加上檔案是否存在的判斷
if os.path.isfile("a.txt"):
    a=open("a.txt","r")
    print(a.readlines())
    a.close()
else:
    print("檔案不存在")
```

#split函式

```python
if os.path.isfile("a.txt"):
    a=open("a.txt","r")
    for line in a:
        fields=line.split(" ") #以空白為切割欄位
        print(fields)
    a.close()
else:
    print("檔案不存在")
```

#附加檔案

```python
a=open("a.txt", "a")
a.write("This is test4\n")
a.close()
```

```python
a=open("a.txt","r")
print(a.readlines())
a.close()
```

#比較a與r+的不同
 a寫在檔案的結尾
 
 r+寫在檔案的開頭

```python
a=open("a.txt","r+")
a.write("This is a test5\n")
a.close()
```

```python
a=open("a.txt","r")
print(a.readlines())
a.close()
```

#刪除檔案

```python
if os.path.isfile("a.txt"):
    os.remove("a.txt")
    print("檔案刪除成功")
else:
    print("找不到檔案")
print(os.path.isfile("a.txt"))
```
